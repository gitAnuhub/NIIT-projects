```python
import json
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv('DS1_C8_V3_ND_Sprint2_Data Analysis Using Python_Dataset.csv')
data
```

```python
genres_data=data['genres'].apply(json.loads)
a=[]
b=[]
for i in range(0,4803):
    b.append(a)
    a=[]
    for j in genres_data[i]:
        a.append(j['name'])
b.pop(0)
gen=pd.Series(b)
g=pd.DataFrame(gen,columns=['Genres'])
data=pd.concat([data,g],axis=1)
```

```python
keywords_data=data['keywords'].apply(json.loads)
a=[]
b=[]
for i in range(0,4803):
    b.append(a)
    a=[]
    for j in keywords_data[i]:
        a.append(j['name'])
b.pop(0)
gen=pd.Series(b)
g=pd.DataFrame(gen,columns=['Keywords'])
data=pd.concat([data,g],axis=1)
```

```python
pcompany_data=data['production_companies'].apply(json.loads)
a=[]
b=[]
for i in range(0,4803):
    b.append(a)
    a=[]
    for j in pcompany_data[i]:
        a.append(j['name'])
b.pop(0)
gen=pd.Series(b)
g=pd.DataFrame(gen,columns=['Production_companies'])
data=pd.concat([data,g],axis=1)
```

```python
pcountry_data=data['production_countries'].apply(json.loads)
a=[]
b=[]
for i in range(0,4803):
    b.append(a)
    a=[]
    for j in pcountry_data[i]:
        a.append(j['name'])
b.pop(0)
gen=pd.Series(b)
g=pd.DataFrame(gen,columns=['Production_countries'])
data=pd.concat([data,g],axis=1)
```

```python
language_data=data['spoken_languages'].apply(json.loads)
a=[]
b=[]
for i in range(0,4803):
    b.append(a)
    a=[]
    for j in language_data[i]:
        a.append(j['name'])
b.pop(0)
gen=pd.Series(b)
g=pd.DataFrame(gen,columns=['Spoken_languages'])
data=pd.concat([data,g],axis=1)
```

```python
data_corr=data['popularity'].corr(data['budget'])


plt.scatter(data['popularity'],data['budget'], color='green')
plt.title('correlation between budget and popularity')
plt.ylim(data['budget'].min(),data['budget'].max())
plt.xlim(data['popularity'].min(),data['popularity'].max())
plt.xlabel("R = "+str(data_corr))
```

```python
data_corr=data['budget'].corr(data['popularity'])


plt.scatter(data['budget'],data['popularity'], color='blue')
plt.title('correlation between budget and popularity')
plt.xlim(data['budget'].min(),data['budget'].max())
plt.ylim(data['popularity'].min(),data['popularity'].max())
plt.xlabel("R = "+str(data_corr))
```

#### Interpretation : The popularity and budget are positively correlated as corelation coefficiect is positive
